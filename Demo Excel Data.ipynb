```python
import numpy as np
import csv
from datetime import date
import datetime

def get_file():
    fields = []
    rows = []
    with open("stock_prices.csv", 'r') as f:
            csvreader = csv.reader(f)
            fields = next(csvreader)
            for row in csvreader: 
                rows.append(row)
    return fields, rows

def conv(value):
    if ',' in value: 
        tmp = ''
        tmp_b = value.split(',')
        for i in tmp_b: 
            tmp += i
        return float(tmp)
    if '/' in value: 
        month = value.split('/')[0]
        if len(month) == 1: 
            month = "0" + month
        day = value.split('/')[1]
        year = value.split('/')[2]
        tmp_date = month + "-" + day + "-" + year
        return datetime.datetime.strptime(tmp_date,"%m-%d-%Y")
    try: 
        return float(value)
    except Exception as e: 
        print(e)

fields, rows = get_file()  
    
weights = [0.0, 0.1, 0.3, 0.2, 0.3, 0.5]


# take dot product of the two inputs
def w_sum(a, b):
    output = 0
    assert(len(a) == len(b))
    for i in range(0, len(a)):
        if isinstance(a[i], datetime.datetime):
            dd = int(a[i].timestamp())
            dd = (b[i] * dd)
            output += dd
            continue
        output += a[i] * b[i]
    return output

def neural_network(input, weights):
    predict = w_sum(input, weights)
    return predict


for i in rows:
    input = [conv(t) for t in i]
    predict = neural_network(input, weights)
    print(predict)
    

```
