```python
# gradient descent w/ single input (review)

## here we just want to review the basic concepts surrounded with gradient descent
## and the calculus of optimizing for adjusted weight measures. 

import matplotlib.pyplot as plt

weight = 3
input = 2
alpha = 0.1
goal_pred = 0.3

predictions = []

for i in range(30): 
    # make the initial predictions
    pred = input * weight
    predictions.append(pred)
    
    # calculate the error value
    error = (pred - goal_pred) ** 2
    
    # calculate the weight delta and its derivative
    delta = pred - goal_pred
    derivative = delta * input
    
    # adjust the weight accordingly
                       # multiply by alpha to make large deltas smaller
    weight = weight - (alpha * derivative) 
    
plt.plot(predictions)
plt.show()
```

```python
# Gradient descent learning w/ Multiple Inputs

# dot product
def w_sum(a, b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output

# scalar multiplication (in this case, we are taking the dy/dx value)
def ele_mul(a_value, vector):
    output = [0 for i in range(len(vector))]
    assert(len(vector) == len(output))
    for i in range(len(vector)): 
        output[i] = (a_value) * (vector[i])
    return output

# various weights for each input
weights = [
    0.1, 0.2, -0.1
]

def neural_network(input, weights):
    pred = w_sum(input, weights)
    return pred

# various inputs
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1,1,0,1] # these are the actual results (goal_pred)

# make a single input
input = [toes[0], wlrec[0], nfans[0]]


for i in range(5):
    
    # find the prediction
    pred = neural_network(input, weights)

    # determine the error value.
    error = (pred - win_or_lose_binary[0]) ** 2

    # find the delta of the actual prediction and the real value.
    # ** the delta is the same for each input to weight relationship related to a single output node
    delta = pred - win_or_lose_binary[0]

    # calculate the weight derivative fo.
    weight_deltas = ele_mul(delta, input)

    alpha = 0.01

    for i in range(len(weights)):
        weights[i] = weights[i] - (weight_deltas[i] * alpha)
    
    print("Weights: "+ str(weights))
    print("Pred: " + str(pred))


```
