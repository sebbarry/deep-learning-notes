```python
"""
Hot and Cold Error Modification
"""
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001 # how much to move the weights in each direction

for iteration in range(1200):
    prediction = input * weight
    error = (goal_prediction - prediction)
    error = error ** 2
    
    print("Error: " + str(error) + "Prediction: " + str(prediction))
    
    up_prediction = input * (weight + step_amount)
    down_prediction = input * (weight - step_amount)
    up_error = (goal_prediction - up_prediction) ** 2
    down_error = (goal_prediction - down_prediction) ** 2
    
    if(down_error < up_error): 
        weight = weight - step_amount # if down error is lower, subtract the step_amount from the weight.
    if(up_error < down_error): 
        weight = weight + step_amount # if up error is lower, add the step_amount to the weight.
        
"""
This is inneficient. 

1. It cycles many times 
2. The step amount may never be entirely perfect, and make our predictions hop back and forth w/o sometimes every being 
able to reach out target weight. 
3. The error may be much bigger than the next, however the step amount will always remain staticly small. 
"""
```

```python
# defining the direction and the amount to adjust the step (gradient descent)
weight = 0.5
goal_pred = 5
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input # this becomes the step amount
    weight = weight - direction_and_amount
    print("Error: " +  str(error) + " Prediction: " + str(pred))
```

```python
"""
Example use of alpha and delta values.
"""
weight = 1
alpha = 0.01
def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]
input = number_of_toes[0]
goal_pred = win_or_lose_binary[0]
pred = neural_network(input, weight)

error = (pred - goal_pred) ** 2 # error rate
delta = pred - goal_pred # this is a measure by how much the node missed. This is the raw amounw that the node was too high
# or too low. We first calculate how much you want the output node to be different.
weight_delta = delta * input
 
weight_delta -= weight_delta * alpha # this number is used before updating weight. This controls the speed of the 
# learning process so as not to overshoot the weight adjustments.
```

```python
"""
Delta and Alpha Adjustments in Gradient Descent.
"""
import matplotlib.pyplot as plt
errors = []
predictions = []
goal_pred_graph = []

weight, goal_pred, input = (0.0, 0.8, 0.5)
for iteration in range(20):
    pred = input * weight
    error = ((input * weight) - goal_pred) ** 2
    
    delta = pred - goal_pred
    weight_delta = delta * input
    weight = weight - weight_delta
    print("Error: " + str(error) + " Prediction: " + str(pred))
    
    #plot examples
    errors.append(error)
    predictions.append(pred)
    goal_pred_graph.append(goal_pred)
    
    
plt.plot(errors)
plt.ylabel('Error Rate')
plt.plot(predictions)
plt.xlabel("Predictions")
plt.plot(goal_pred_graph)
plt.show()
```

```python
# calculus with gradient descent (relabeled)
import matplotlib.pyplot as plt

weight = 0.0
goal_pred = 0.8
input = 1.1

points = []

for iteration in range(20):
    prediction = input * weight
    error = (prediction - goal_pred) ** 2 # mean squared to find the error value
    d_delta = (prediction - goal_pred) * input # dx/dy of the prediction in rleation to error
    weight = weight - d_delta # change the weight value accordingly. 
    # if the d_delta is negative... weight - (-d_delta) ~ weight + d_delta
    print("Error: " + str(error) + " Prediction: " + str(prediction))
    points.append(prediction)

plt.plot(points)
plt.show()
```

```python
# introduction of alpha (to minimize the divergence of a network if an input is large)
import matplotlib.pyplot as plt


weight = 0.5
goal_pred = 0.8
input = 2
alpha = 0.1

predictions = []
weights = []

for iteration in range(10):
    prediction = input * weight
    error = (prediction - goal_pred) ** 2
    
    d_delta = (prediction - goal_pred) * input
    weight = weight - (alpha * d_delta)
    weights.append(weight)
    print("Error: " + str(error) + " Prediction: " + str(prediction))
    predictions.append(prediction) # notice in the graph, there is little dampening.
    
plt.plot(predictions)
plt.show()
```
